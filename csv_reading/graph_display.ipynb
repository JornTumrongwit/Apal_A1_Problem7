```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def display(slist, built):
    df = pd.read_csv(slist)
    df_map = pd.read_csv(built)
    fig, ax = plt.subplots()
    ax.plot(df["list_size"], df["time"], label="skip list")
    ax.plot(df_map["list_size"], df_map["time"], label="built-in")
    ax.legend()
    fig.show()
```

```python
display('../experiments/insert.csv', '../experiments/insert_map.csv')
```

```python
display('../experiments/search.csv', '../experiments/search_map.csv')
```

```python
display('../experiments/erase.csv', '../experiments/erase_map.csv')
```

```python
def display_special(slist, other_slist, built):
    df = pd.read_csv(slist)
    df_other = pd.read_csv(other_slist)
    df_map = pd.read_csv(built)
    fig, ax = plt.subplots()
    ax.plot(df["list_size"], df["time"], label="skip list")
    ax.plot(df_other["list_size"], df_other["time"], label="skip list (bi-directional search)")
    ax.plot(df_map["list_size"], df_map["time"], label="built-in")
    ax.legend()
    fig.show()
```

```python
display_special('../experiments/search.csv', '../experiments/search_bidir.csv', '../experiments/search_map.csv')
```

```python
def display_space(slist, built):
    df = pd.read_csv(slist)
    df_map = pd.read_csv(built)
    fig, ax = plt.subplots()
    ax.plot(df["size"], df["mem_MB"], label="skip list")
    ax.plot(df_map["size"], df_map["mem_MB"], label="built-in")
    ax.legend()
    fig.show()
```

```python
display_space('../experiments/space.csv', '../experiments/space_map.csv')
```
